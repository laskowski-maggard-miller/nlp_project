```python
import pandas as pd
import numpy as np
import re
import requests
from bs4 import BeautifulSoup as bs

from time import sleep
```

```python
url = 'https://github.com/search?q=cats&type=repositories'
r = requests.get(url)
soup = bs(r.content, 'html.parser')
```

# Finding repos:

#### Finds all repos on a single repo results page (10 results):

```python
# finds all repo urls on the page
repo_list = []
repo = soup.find('ul', class_ = "repo-list")
repos = repo.find_all('a', class_ = 'v-align-middle')
for thing in repos:
    repo_url = thing.text
    repo_list.append(repo_url)
```

Pages 1 - 8

```python
start_page = 1
end_page = 9

for page in range(start_page,end_page):
    url = f'https://github.com/search?p={page}&q=cats&type=Repositories'
    print(f'Procuring {url}')
    r = requests.get(url)
    soup = bs(r.content, 'html.parser')
    
    repo = soup.find('ul', class_ = "repo-list")
    try:
        repos = repo.find_all('a', class_ = 'v-align-middle')
    except:
        print(f'>>>Failed to procure {url}')
        break
    for thing in repos:
        repo_url = thing.text
        repo_list.append(repo_url)
    sleep(2)
```
